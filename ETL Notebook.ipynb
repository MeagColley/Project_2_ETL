```python
# some dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# The path to our CSV files
file1 = "Data/honeyproduction.csv"
file2 = "Data/vHoneyNeonic_v02.csv"
file3 = "Data/vHoneyNeonic_v03.csv"
```

```python
# make database 1 a dataframe
honeydf = pd.read_csv(file1)
honeydf.head()
```

```python
# take csv from database 2 and make it a datafram
pesticidedf1 = pd.read_csv (file2)
pesticidedf1
```

```python
# take second csv from database 2 and make it a dataframe
pesticidedf2 = pd.read_csv (file3)
pesticidedf2
```

```python
# combine the two dataframes that are the database 2 to make one dataframe for database 2
# pesticide_df = pd.merge(pesticidedf1,pesticidedf2,  how="outer")
# pesticide_df
pesticidedf1 = pesticidedf1.append(pesticidedf2, sort=False)
pesticidedf1
```

```python
# because honeydf goes all the way to 2012 and pesticide df only goes to 2007, lets drop from honeydf all years following 2007
# capture the years to drop in a variable 
yearsdrop = honeydf[ honeydf['year'] > 2007 ].index
# Delete these row indexes from dataFrame
honeydf.drop(yearsdrop , inplace=True)
honeydf
```

```python
# now lets drop columns from pesticide_df that don't appear in honey_df
pesticidedf1.drop(['StateName', 'Region', 'FIPS'], axis = 1)
```
